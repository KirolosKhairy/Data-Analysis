##🧠 Часть 3 — Оптимизация типов в pandas (Housing.csv)

🔧 Установка и импорт

```python
import pandas as pd
import numpy as np
```

###📥 Чтение датасета

```python
df = pd.read_csv("/content/Housing.csv")
df.head()
```

###📦 Базовое потребление памяти

```python
base_mem_bytes = df.memory_usage(deep=True).sum()
base_mem_mb = base_mem_bytes / (1024**2)
pd.DataFrame({"metric":["memory_before_mb"],"value":[base_mem_mb]})
```

###🧭 Автоподбор оптимальных типов

```python
from pandas.api.types import is_integer_dtype, is_float_dtype, is_bool_dtype, is_object_dtype

def suggest_dtype(s):
    name = s.name
    non_null = s.notna().sum()
    nunique = s.nunique(dropna=True)
    pct_unique = (nunique / non_null) if non_null else np.nan
    orig = str(s.dtype)
    sugg = orig
    reason = "keep"
    if is_integer_dtype(s):
        s2 = pd.to_numeric(s, downcast="integer")
        sugg = str(s2.dtype)
        reason = "downcast_integer"
    elif is_float_dtype(s):
        s2 = pd.to_numeric(s, downcast="float")
        sugg = str(s2.dtype)
        reason = "downcast_float"
    elif is_bool_dtype(s):
        if s.isna().any():
            sugg = "boolean"
        else:
            sugg = "bool"
        reason = "bool"
    elif is_object_dtype(s):
        s_dt = pd.to_datetime(s, errors="coerce", infer_datetime_format=True)
        ok = s_dt.notna().sum()
        if non_null and ok / non_null >= 0.95 and ok > 0:
            sugg = "datetime64[ns]"
            reason = "datetime"
        else:
            vals = s.dropna().astype(str).str.lower().str.strip()
            m_true = {"true","yes","y","1"}
            m_false = {"false","no","n","0"}
            is_boolish = vals.isin(m_true.union(m_false))
            if non_null and is_boolish.mean() >= 0.95:
                sugg = "boolean"
                reason = "bool_from_text"
            else:
                if non_null and nunique > 0 and (nunique <= max(50, int(0.5*non_null))):
                    sugg = "category"
                    reason = "low_cardinality"
                else:
                    sugg = "object"
                    reason = "text"
    return orig, sugg, non_null, nunique, pct_unique, reason

profile = []
for c in df.columns:
    orig, sugg, non_null, nunique, pct_unique, reason = suggest_dtype(df[c])
    profile.append([c, orig, sugg, non_null, nunique, pct_unique, reason])

report = pd.DataFrame(profile, columns=["column","orig_dtype","suggested_dtype","non_null","n_unique","pct_unique","reason"])
report
```

###🔄 Применение оптимизаций типов

```python
df_opt = df.copy()
for _, row in report.iterrows():
    c = row["column"]
    target = row["suggested_dtype"]
    if target == row["orig_dtype"]:
        continue
    if target.startswith("int") or target.startswith("uint"):
        df_opt[c] = pd.to_numeric(df_opt[c], downcast="integer")
    elif target.startswith("float"):
        df_opt[c] = pd.to_numeric(df_opt[c], downcast="float")
    elif target == "boolean":
        try:
            df_opt[c] = df_opt[c].astype("boolean")
        except:
            vals = df_opt[c].astype(str).str.lower().str.strip()
            df_opt[c] = vals.replace({"true":True,"yes":True,"y":True,"1":True,"false":False,"no":False,"n":False,"0":False}).astype("boolean")
    elif target == "bool":
        df_opt[c] = df_opt[c].astype(bool)
    elif target == "datetime64[ns]":
        df_opt[c] = pd.to_datetime(df_opt[c], errors="coerce", infer_datetime_format=True)
    elif target == "category":
        df_opt[c] = df_opt[c].astype("category")
```

###⚖️ Сравнение памяти: до vs после

```python
opt_mem_bytes = df_opt.memory_usage(deep=True).sum()
opt_mem_mb = opt_mem_bytes / (1024**2)
saving_mb = base_mem_mb - opt_mem_mb
saving_pct = (saving_mb / base_mem_mb * 100) if base_mem_mb else np.nan
pd.DataFrame({
    "metric":["memory_before_mb","memory_after_mb","saving_mb","saving_pct"],
    "value":[base_mem_mb, opt_mem_mb, saving_mb, saving_pct]
})
```

###🗂️ Изменения по столбцам

```python
mem_before = df.memory_usage(index=False, deep=True)
mem_after = df_opt.memory_usage(index=False, deep=True)
delta = (mem_before - mem_after).sort_values(ascending=False)
changes = report.loc[report["orig_dtype"] != report["suggested_dtype"], ["column","orig_dtype","suggested_dtype","reason"]].set_index("column")
changes = changes.join(pd.DataFrame({"mem_before":mem_before, "mem_after":mem_after, "mem_saved":delta}))
changes.sort_values("mem_saved", ascending=False)
```

###✅ Проверка диапазонов и типов до/после

```python
import pandas as pd, numpy as np
num_cols = ["price","area","bedrooms","bathrooms","stories","parking"]
def rng_tbl(d, cols):
    return pd.DataFrame({"min": d[cols].min(), "max": d[cols].max(), "dtype": d[cols].dtypes.astype(str)})
rng_before = rng_tbl(df, num_cols)
rng_after = rng_tbl(df_opt, num_cols).add_prefix("after_")
rng_compare = rng_before.join(rng_after)
rng_compare
```
